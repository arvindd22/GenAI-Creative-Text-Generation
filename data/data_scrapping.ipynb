```python
import requests
from bs4 import BeautifulSoup
import time

def scrape_poems(url , poems):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        rows = soup.find_all('article')
        if not rows:
            print("No articles found with the specified class.")
            return
        for row in rows:
            title_cell = row.find('a', class_="link-underline-on link-red")
            poet_cell = row.find('div', class_="type-kappa text-gray-600")
            if title_cell and poet_cell:

                title = title_cell.find('span').text.strip()
                poem_url = "https://www.poetryfoundation.org" + title_cell['href']
                poet = poet_cell.find('span').text.strip()

                print(f"Title: {title}, Poem URL: {poem_url}, Poet: {poet}")
                poems.append([title, poem_url, poet])

            # Uncomment the line below during debugging if needed
            # time.sleep(1)
    else:
        print(f"Failed to retrieve page: {response.status_code}")

poems = []
scrape_poems("https://www.poetryfoundation.org/poems/browse" , poems)

```

```python
from bs4 import BeautifulSoup
import requests

def scrape_title_column_links(base_url, num_pages):
  """
  Scrapes poem link data from all pages (up to the specified number)
  starting from the given base URL.

  Args:
    base_url: The base URL of the website containing poem links.
    num_pages: The number of pages to scrape (maximum 785).

  Returns:
    A list of dictionaries, where each dictionary holds the poem title
    and its corresponding link.
  """

  all_links = []
  for page_number in range(0, num_pages - 1):
    url = f"{base_url}/{page_number}"  # Construct URL for each page

    try:
      response = requests.get(url)
      response.raise_for_status()  # Raise an exception for bad status codes
      soup = BeautifulSoup(response.content, 'html.parser')

      # Find the table element (adjust the selector as needed)
      table = soup.find('table')  # Example: Find the first table

      if table:
        links = []
        # Assuming the title column is the first column (index 0)
        for row in table.find_all('tr'):
          try:
            title_cell = row.find_all('th')[0]  # Get the first cell (title column)
          except IndexError:
            title_cell = row.find_all('td')[0]  # If no 'th', try 'td'

          for link_tag in title_cell.find_all('a'):
            link = f"https://poets.org{link_tag.get('href')}"
            name = link_tag.get_text(strip=True)  # Extract link text
            links.append({'link': link, 'name': name})

        all_links.extend(links)  # Add scraped links to the overall list
      else:
        print(f"No table found on page: {url}")

    except requests.exceptions.RequestException as e:
      print(f"Error fetching URL {url}: {e}")

  return all_links

# Example usage
base_url = "https://poets.org/poems?page="
num_pages = 501  # Adjust the number of pages to scrape (maximum 785)

all_poem_links = scrape_title_column_links(base_url, num_pages)

if all_poem_links:
  # Process the scraped data (all_poem_links)
  print("Successfully scraped links from all pages!")
else:
  print("No links found during scraping.")
```

```python
from bs4 import BeautifulSoup
import requests

def scrape_poem(url):
  """
  Scrapes the poem from the given URL.

  Args:
    url: The URL of the webpage containing the poem.

  Returns:
    The poem as a string, or None if the poem could not be found.
  """

  try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the div containing the poem
    poem_div = soup.find('div', class_='field field--body')

    if poem_div:
      # Extract the poem text
      poem_text = poem_div.get_text(strip=True)
      return poem_text
    else:
      print(f"Div with class 'field field--body' not found on {url}")
      return None

  except requests.exceptions.RequestException as e:
    print(f"Error fetching URL: {e}")
    return None

```

```python
import json
import re

# Load raw data from JSON file (replace 'raw_poems.json' with your file name)
with open("raw_poems.json", "r") as file:
    raw_data = json.load(file)

# Function to clean individual poem content
def clean_poem(poem_text):
    # Remove non-printable and special characters, except essential punctuation
    poem_text = re.sub(r"[^\w\s,.!?;:'\-]", "", poem_text)
    # Normalize spaces (remove extra spaces, tabs, etc.)
    poem_text = re.sub(r"\s+", " ", poem_text).strip()
    return poem_text

# Cleansing process
cleansed_data = []
for entry in raw_data:
    if "poem" in entry and entry["poem"].strip():
        cleaned_entry = {
            "title": entry.get("title", "Untitled"),  # Default title if missing
            "poem": clean_poem(entry["poem"]),
        }
        cleansed_data.append(cleaned_entry)

# Remove duplicates based on poem content
unique_data = []
seen_poems = set()
for entry in cleansed_data:
    if entry["poem"] not in seen_poems:
        unique_data.append(entry)
        seen_poems.add(entry["poem"])

# Save cleansed data to a new JSON file
with open("cleansed_poems.json", "w") as file:
    json.dump(unique_data, file, indent=4)

# Save cleansed poems to a text file
with open("cleansed_poems.txt", "w") as file:
    for entry in unique_data:
        file.write(f"Title: {entry['title']}\n")
        file.write(f"{entry['poem']}\n\n")

print(f"Cleansed dataset contains {len(unique_data)} unique entries.")
```
