```python
from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
input = "This is a basic example to demonstrate working of the tokenizers!"
```

```python
tokens = tokenizer.tokenize(input)
print(tokens)
```

what is that weird symbol before every token?

```python
encoded = tokenizer.encode(input)
print(encoded)
```

```python
# Now we introduce some special token, EOD, BOS, and PAD
tokenizer.add_special_tokens({'pad_token': '[PAD]','bos_token': '[BOS]','eos_token': '[EOS]'})
```

```python
input = "[BOS]" + input + "[EOS]"
print(input)
```

```python
tokens = tokenizer.tokenize(input)
print(tokens)
```

```python
encoded = tokenizer.encode(input, padding = 'max_length', max_length = 20)
print(encoded)
```

```python
tokenizer.decode(encoded)
```

```python
# also, for extra long sentences, we can truncate them to fit to max_length
input = 'this is an example of an extra long sentence!'
encoded = tokenizer.encode(input, truncate = 'max_length', max_length = 5)
print(encoded)
```

```python
print(tokenizer.decode(encoded))
```

```python
# in practicality, we deal with a batch of inputs, not a singular input
# and we want them to be of a uniform length, that's why we use padding, and truncation
# and to separate different entities (like two small poems), we add EOS and BOS tokens
# here is an example to deal with batch inputs
example = ['this is a', 'wonderful', 'example to demonstrate the importance of using', 'special tokens', 'when we deal with batch inputs']
encoded = tokenizer(example, padding = 'max_length', truncation = True , max_length = 6)
for i, seq in enumerate(example):
  print(f'for seq {i}')
  print('tokens: ',encoded['input_ids'][i])
  print('attention mask:',encoded['attention_mask'][i])
```

```python
#one last bit
#you can enable the tokenizer to adapt to the type of model used directly
#AutoTokenizer helps us pick the tokenizer automatically based on the model used
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
sequence = "Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)
print(tokens)
tokenizer
```
